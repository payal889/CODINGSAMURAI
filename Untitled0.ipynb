```python
import pandas as pd
import matplotlib.pyplot as plt

movies=pd.read_csv("movies.csv")
ratings=pd.read_csv("ratings.csv")

movies.head()
ratings.head()

avg_rating=ratings.groupby("movieId")['rating'].mean()
avg_rating.head()

top_movies=avg_rating.sort_values(ascending=False).head(10)
top_movies

data=pd.merge(ratings,movies,on='movieId')
data.head()

movies['genres'] = movies['genres'].str.split('|')

genre_count = movies.explode('genres')['genres'].value_counts()

genre_count.plot(kind='bar')
plt.title("Movies per Genre")
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.show()

comedy = data[data['genres'].str.contains("Comedy")]

recommend = comedy.groupby('title')['rating'].mean().sort_values(ascending=False).head(5)

recommend

```
